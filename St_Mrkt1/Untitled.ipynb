```python
#Training and Testing( 7:3)
d_train=pd.DataFrame(df['Close'][0:int(len(df)*0.70)])
d_test=pd.DataFrame(df['Close'][int(len(df)*0.70):])
```

```python
d_test.shape
```

```python
d_train.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
# feature will be ranged b/w 0-1
scaler=MinMaxScaler(feature_range=(0,1))
```

```python
d_train_array=scaler.fit_transform(d_train)
```

```python
d_train_array
```

```python
d_train_array.shape
```

```python
d_train_array[109-100:109]
```

```python
d_train_array[100,0]
d_train_array[100][0]
```

```python
x_train=[]
y_train=[]
for i in range(100,d_train_array.shape[0]):
    x_train.append(d_train_array[i-100:i])
    y_train.append(d_train_array[i,0])
x_train, y_train  = np.array(x_train), np.array(y_train)    
```

```python
x_train
```

```python
x_train.shape
```

```python
x_train.shape[1]
```

```python
x_train[0][0]
```

```python
#Model building
from keras.layers import Dropout,Dense,LSTM
from keras.models import Sequential
```

```python
model = Sequential()
model.add(LSTM(units=50,activation='relu',return_sequences = True, input_shape = (x_train.shape[1],1)))
model.add(Dropout(0.2))

model.add(LSTM(units = 60, activation = 'relu', return_sequences = True))
model.add(Dropout(0.3))

model.add(LSTM(units = 80, activation = 'relu', return_sequences = True))
model.add(Dropout(0.4))

model.add(LSTM(units = 120, activation = 'relu'))
model.add(Dropout(0.5))

model.add(Dense(units = 1))
```

```python
model.summary()
```

```python
model.compile(optimizer = 'adam', loss = 'mean_squared_error')
model.fit(x_train, y_train, epochs = 50)
```

```python
past_100_days = d_train.tail(100)
```

```python
past_100_days
```

```python
# we have added last 100 because for calculating new result on the d_test data we need last 100 as we are making prdiction based on last 100
final_df = pd.concat([past_100_days, d_test], ignore_index=True)
```

```python
final_df.head()
```

```python
input_data = scaler.fit_transform(final_df)
```

```python
input_data[2:5]
```

```python
x_test = []
y_test = []

for i in range(100, input_data.shape[0]):
    #last 100 values before i
    x_test.append(input_data[i-100:i])
    #actual value of that i
    y_test.append(input_data[i, 0])

x_test, y_test  = np.array(x_test), np.array(y_test)
```

```python
x_test.shape
```

```python
y_predicted = model.predict(x_test)
```

```python
y_predicted.shape
```

```python
scaler.scale_
```

```python
scaler_factor = 1 / 0.00339369
y_predicted = y_predicted * scaler_factor
y_test = y_test * scaler_factor
```

```python
plt.figure(figsize=(12, 6))
plt.plot(y_test, label = 'Original Price', linewidth = 1)
plt.plot(y_predicted, label = 'Predicted Price', linewidth = 1)
plt.legend()
plt.show()
```
